```python
import numpy as np
import objectives
import solvers
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rcParams.update({'font.size': 15})
```

```python
rosenbrock_2 = objectives.get_rosenbrock(2)
rosenbrock_5 = objectives.get_rosenbrock(5)
rosenbrock_10 = objectives.get_rosenbrock(10)
rosenbrock_50 = objectives.get_rosenbrock(50)
```

```python
c0 = .4
c1 = .5
tol = 1e-6
```

```python
#n = 2
n=2
sd = solvers.steepest_descent(rosenbrock_2[-1], rosenbrock_2[0], rosenbrock_2[1], c0, c1, 1, tol)
newton = solvers.newton(rosenbrock_2[-1], rosenbrock_2[0], rosenbrock_2[1], rosenbrock_2[2], c0, c1, 1, tol)
bfgs = solvers.BFGS(rosenbrock_2[-1], rosenbrock_2[0], rosenbrock_2[1], c0, c1, 1, tol)
dfp = solvers.DFP(rosenbrock_2[-1], rosenbrock_2[0], rosenbrock_2[1], c0, c1, 1, tol)
```

```python
#n = 5
n=5
sd = solvers.steepest_descent(rosenbrock_5[-1], rosenbrock_5[0], rosenbrock_5[1], c0, c1, 1, tol)
newton = solvers.newton(rosenbrock_5[-1], rosenbrock_5[0], rosenbrock_5[1], rosenbrock_5[2], c0, c1, 1, tol)
bfgs = solvers.BFGS(rosenbrock_5[-1], rosenbrock_5[0], rosenbrock_5[1], c0, c1, 1, tol)
dfp = solvers.DFP(rosenbrock_5[-1], rosenbrock_5[0], rosenbrock_5[1], c0, c1, 1, tol)
```

```python
#n = 10
n=10
sd = solvers.steepest_descent(rosenbrock_10[-1], rosenbrock_10[0], rosenbrock_10[1], c0, c1, 1, tol)
newton = solvers.newton(rosenbrock_10[-1], rosenbrock_10[0], rosenbrock_10[1], rosenbrock_10[2], c0, c1, 1, tol)
bfgs = solvers.BFGS(rosenbrock_10[-1], rosenbrock_10[0], rosenbrock_10[1], c0, c1, 1, tol)
dfp = solvers.DFP(rosenbrock_10[-1], rosenbrock_10[0], rosenbrock_10[1], c0, c1, 1, tol)
```

```python
#n = 50
n=50
sd = solvers.steepest_descent(rosenbrock_50[-1], rosenbrock_50[0], rosenbrock_50[1], c0, c1, 1, tol)
newton = solvers.newton(rosenbrock_50[-1], rosenbrock_50[0], rosenbrock_50[1], rosenbrock_50[2], c0, c1, 1, tol)
bfgs = solvers.BFGS(rosenbrock_50[-1], rosenbrock_50[0], rosenbrock_50[1], c0, c1, 1, tol)
dfp = solvers.DFP(rosenbrock_50[-1], rosenbrock_50[0], rosenbrock_50[1], c0, c1, 1, tol)
```

```python
fig, ax = plt.subplots()
ax.set_yscale('log')
ax.set_title(f'n={n} grad norms vs. iterations')

ax.plot(np.indices((len(sd[-1]),))[0], sd[-1], label="steepest descent grad norms")
ax.plot(np.indices((len(newton[-1]),))[0], newton[-1], label="newton grad norms")
ax.plot(np.indices((len(bfgs[-1]),))[0], bfgs[-1], label="bfgs grad norms")
ax.plot(np.indices((len(dfp[-1]),))[0], dfp[-1], label="dfp grad norms")

ax.axhline(y = tol, color = 'black', linestyle='--', label='tolerance')

ax.legend(loc='lower center')
```

```python
fig, ax = plt.subplots()
ax.set_yscale('log')
ax.set_title(f'n={n} function values vs. iterations')

ax.plot(np.indices((len(sd[1]),))[0], sd[1], label="steepest descent function values")
ax.plot(np.indices((len(newton[1]),))[0], newton[1], label="newton function values")
ax.plot(np.indices((len(bfgs[1]),))[0], bfgs[1], label="bfgs function values")
ax.plot(np.indices((len(dfp[1]),))[0], dfp[1], label="dfp function values")

ax.legend(loc='lower center')
```

```python
print('Total running time for each algorithm:')
print(f'SD: {round(sd[2][-1]*1000, 3)}ms')
print(f'Newton: {round(newton[2][-1]*1000,3)}ms')
print(f'DFP: {round(dfp[2][-1]*1000,3)}ms')
print(f'BFGS: {round(bfgs[2][-1]*1000,3)}ms')
```
